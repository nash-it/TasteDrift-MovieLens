```python
#Genre distribution
import matplotlib.pyplot as plt
df['genres'] = df['genres'].str.split('|')
df_exploded = df.explode('genres')

plt.figure(figsize = (10,5))
df_exploded['genres'].value_counts().head(30).plot(kind = 'bar')
plt.title("Top 10 Generes by Rating Count")
plt.xlabel("Genres")
plt.ylabel("Rating Count")
plt.tight_layout()
plt.show()
```

```python
df[df['user_id']==1].sort_values('datetime')
```

```python
# Ratings per year

ratings_per_year = df['datetime'].dt.year.value_counts().sort_index()
plt.figure(figsize = (10,4))
ratings_per_year.plot(kind = 'bar', title= "Ratings Counts by Year", color='skyblue')
plt.ylabel("number of Ratings")
plt.xlabel("Year")
plt.tight_layout()
plt.show()
```

```python
# Taste timeline: early vs late

df['user_order'] = df.groupby('user_id')['datetime'].rank(method = 'first')
df['user_total'] = df.groupby('user_id')['user_order'].transform('max')
df['period'] = df['user_order'] / df['user_total']
df['time_split'] = df['period'].apply(lambda x: 'late' if x>0.5 else 'early')
split_df = df.copy()
# if split_df['genres'].apply(lambda x: isinstance(x, list)).all():
#     split_df['genres'] = split_df['genres'].explode().reset_index(drop=True)
# else:
#     logger.warning("Some entries in 'genres' are not lists. Attempting to split them first.")
#     split_df['genres'] = split_df['genres'].apply(lambda g: g.split('|') if isinstance(g, str) else g)
#     split_df['genres'] = split_df['genres'].explode()

# split_df['genres'] = split_df['genres'].apply(lambda g: g if isinstance(g, list) else g.split('|') if isinstance(g, str) else [])
# split_df['genres'] = split_df['genres'].explode()

import pandas as pd

split_df['genres'] = split_df['genres'].apply(
    lambda g: g if isinstance(g, list)
    else g.split('|') if isinstance(g, str)
    else [] if not pd.isna(g)
    else []
)

split_df['genres'] = split_df['genres'].explode().reset_index(drop=True)


taste_shift_summary = split_df.groupby('time_split')['genres'].value_counts().unstack().fillna(0)

taste_shift_summary

taste_diff = taste_shift_summary.loc['late'] - taste_shift_summary.loc['early']
# taste_diff
# top_shifted_genres = taste_diff.abs().sort_values(ascending =False).head(10)
top_shifted_genres = taste_diff.sort_values(key=lambda x: x.abs(), ascending=False).head(10)
# top_shifted_genres

plt.figure(figsize=(10,5))
# top_shifted_genres.plot(kind='barh', color = 'coral')
top_shifted_genres.plot(kind='barh', color=top_shifted_genres.apply(lambda x: 'green' if x > 0 else 'red'))
plt.title("Top Genres with Most Taste Drift(Late vs Early period)")
plt.xlabel("Change in Rating Count")
plt.tight_layout()
plt.show()
```
